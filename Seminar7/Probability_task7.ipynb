```python
import numpy as np
from scipy import stats as st
```

```python
#Задача 1
#Выберете тест и проверьте, есть ли различия между выборками:
#Даны две независимые выборки. Не соблюдается условие нормальности
#x1 380,420, 290
#y1 140,360,200,900
#Сделайте вывод по результатам, полученным с помощью функции
```

```python
#Решение

def compare (p,alpha):
    if p<alpha:
        print("Отвергаем нулевую гипотезу")
    else:
        print("Принимаем нулевую гипотезу")

x1 = np.array([380,420, 290])
y1 = np.array([140,360,200,900])
#По условию задачи, сравниваем 2 независымые выборки,
#для сравнения будем использовать критерий Манна-Уитни
#Нулевая гипотеза заключается в том, что различия есть
#Альтернативная - различий нет
result1 = st.mannwhitneyu(x1,y1)
print(result1)
compare(result1[1],0.05)
```

```python
#Задача 2
#Исследовалось влияние препарата на уровень давления пациентов. 
#Сначала измерялось давление до приема препарата, потом через 10 минут и через 30 минут. 
#Есть ли статистически значимые различия?
#1е измерение до приема препарата: 150, 160, 165, 145, 155
#2е измерение через 10 минут: 140, 155, 150, 130, 135
#3е измерение через 30 минут: 130, 130, 120, 130, 125
```

```python
#Решение
x2 = np.array([150, 160, 165, 145, 155])
y2 = np.array([140, 155, 150, 130, 135])
z2 = np.array([130, 130, 120, 130, 125])
#Из условия задачи видим множественные сравнения повторных измерений,
#используем критерий Фридмана
result2 = st.friedmanchisquare(x2,y2,z2)
print(result2)
compare(result1[1],0.05)
```

```python
#Задача 3
#Сравните 1 и 2 е измерения , предполагая, что 3го измерения через 30 минут не было.
x3 = np.array([150, 160, 165, 145, 155])
y3 = np.array([140, 155, 150, 130, 135])
#По условию, сравниваем две зависимые выборки
#Используем критерий Уилкокосона
result3 = st.wilcoxon(x3,y3)
print(result3)
compare(result3[1],0.05)
```

```python
#Задача 4
#Даны 3 группы учеников плавания.
#В 1 группе время на дистанцию 50 м составляют:
#56, 60, 62, 55, 71, 67, 59, 58, 64, 67
#Вторая группа : 57, 58, 69, 48, 72, 70, 68, 71, 50, 53
#Третья группа: 57, 67, 49, 48, 47, 55, 66, 51, 54
```

```python
#Решение
x4 = np.array([56, 60, 62, 55, 71, 67, 59, 58, 64, 67])
y4 = np.array([57, 58, 69, 48, 72, 70, 68, 71, 50, 53])
z4 = np.array([57, 67, 49, 48, 47, 55, 66, 51, 54])
#По условию сравниваем три независимые выборки
#Используем критерий Крусколла-Уоллеса
result4 = st.kruskal(x4,y4,z4)
print(result4)
compare(result4[1],0.05)
```

```python
#Задача 5
#Заявляется, что партия изготавливается со средним арифметическим 2,5 см. 
#Проверить данную гипотезу, если известно, что размеры изделий подчинены нормальному закону распределения. 
#Объем выборки 10, уровень статистической значимости 5%
#2.51, 2.35, 2.74, 2.56, 2.40, 2.36, 2.65, 2.7, 2.67, 2.34

x5 = np.array([2.51, 2.35, 2.74, 2.56, 2.40, 2.36, 2.65, 2.7, 2.67, 2.34])
alpha = 0.05
mu = 2.5 #Среднее по ГС
n = len(x5)#Размер выборки
print("Размер выборки:",n)
#Нулевая гипотеза: mu = mu_0 = 2.5(партия изготавливается со средним ар. = 2.5 см)
#Альтеранативная гипотеза mu не равно m_0 (среднее ар.партии не равно 2ю5 см)
x_avg = np.mean(x5) #Находим среднее по выборке
print("Среднее по выборке:",x_avg)

d_x = np.var(x5,ddof=1) #Расчет несмещенной дисперсии
print("Несмещенная дисперсия:",d_x)    
#В условии задачи не указано стандартное отклонение генеральной совокупности, поэтому применим расчет t-критерия
t_crit = (x_avg - mu)/(np.sqrt(d_x)/np.sqrt(n))
print("Наблюдаемое значение t-критерия:",t_crit)
#Подбираем табличное значение
t_table = 2.26
print("Табличное значение",t_table)

#Проверяем с помощью функции
tt=st.t.ppf(1 - alpha/2,len(x5)-1)
print(tt)

print("Принимаем нулевую гипотезу, так как значение t-наблюдаемого меньше t-табличного")
print(t_crit<tt)
#Проверка встроенной функцией
result5 = st.ttest_1samp(x5,2.5)
compare(result5[1],0.05)
```
