```python
#Задача 2
#О случайной непрерывной равномерно распределенной величине B известно, что ее дисперсия равна 0.2.
#Можно ли найти правую границу величины B и ее среднее значение зная, что левая граница равна 0.5?
#Если да, найдите ее.
```

```python
#Решение
#Зная формулу дисперсии D = ((b-a)**2)/12), вычислим правую границу величины B
a_2 = 0.5
b_2 = Symbol('b_2')
#Решаем уровнение
roots = solve((((b_2-0.5)**2)/12)-0.2)
#Получаем правую границу
print("Значение правой границы величины B: ",round((roots[1]),4))
#Поиск среднего значения
M_x2 = ((a_2 + roots[1])/2)
print("Среднее значение:",round((M_x2),4))
    
```

```python
#Задача 4
#Рост взрослого населения города X имеет нормальное распределение.
#Причем, средний рост равен 174 см, а среднее квадратичное отклонение равно 8 см.
#Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:
#а). больше 182 см
#б). больше 190 см
#в). от 166 см до 190 см
#г). от 166 см до 182 см
#д). от 158 см до 190 см
#е). не выше 150 см или не ниже 190 см
#ё). не выше 150 см или не ниже 198 см
#ж). ниже 166 см.
```

```python
#Решение
#Общая функция для вычисления z-значений
def z_values(x,mu,sigma):
    output_result = (x-mu)/sigma
    return output_result
#Для решения используем функцию cdf модуля scipy
#Кумулятивная функция распределения ( CDF ) говорит нам о вероятности того, 
#что случайная величина примет значение, меньшее или равное некоторому значению.
#Таким образом,для того,чтобы найти вероятность того, 
#что случайная величина примет значение больше заданного, необходимо вычитать это значение из 1.

#Решение а
a_4 =  z_values(182,174,8)
print("Вероятность того, что случайным образом выбранный взрослый человек имеет рост больше 182 см",round((1-norm.cdf(a_4)),4))

#Решение б
b_4 = z_values(190,174,8)
print("Вероятность того, что случайным образом выбранный взрослый человек имеет рост больше 190 см",round((1-norm.cdf(b_4)),4))

#Решение в
c_4 = z_values(166,174,8)
c_4_result = norm.cdf(b_4) - norm.cdf(c_4)
print("Вероятность того, что случайным образом выбранный взрослый человек имеет рост от 166 см до 190 см",round((c_4_result),4))

#Решение г
d_4 = norm.cdf(a_4) - norm.cdf(c_4)
print("Вероятность того, что случайным образом выбранный взрослый человек имеет рост от 166 см до 182 см",round((d_4),4))

#Решение д
e_4 = z_values(158,174,8)
e_4_result = norm.cdf(b_4) - norm.cdf(e_4)
print("Вероятность того, что случайным образом выбранный взрослый человек имеет рост от 158 см до 190 см",round((e_4_result),4))

#Решение e
f_4_1 = z_values(150,174,8)
f_4_2 = z_values(190,174,8)
f_4_result = norm.cdf(f_4_1) + (1 - norm.cdf(f_4_2))
print("Вероятность того, что случайным образом выбранный взрослый человек имеет рост не выше 150 см или не ниже 190 см",round((f_4_result),4))

#Решение ё
g_4_1 = z_values(150,174,8)
g_4_2 = z_values(198,174,8)
g_4_result = norm.cdf(g_4_1) + (1 - norm.cdf(g_4_2))
print("Вероятность того, что случайным образом выбранный взрослый человек имеет рост не выше 150 см или не ниже 198 см",round((g_4_result),4))

#Решение ж
h = z_values(166,174,8)
print("Вероятность того, что случайным образом выбранный взрослый человек имеет рост ниже 166 см",round((norm.cdf(h)),4))
```

```python
#Задача 5
#На сколько сигм (средних квадратичных отклонений) отклоняется рост человека, равный 190 см, от
#математического ожидания роста в популяции, в которой M(X) = 178 см и D(X) = 25 кв.см?
```

```python
#Решение
#Для решения задачи, воспользуемся формулой Z = (x-mu)/sigma
result_5 = (190 - 178)/(sqrt(25))
print("Рост человека отклоняется на",float(result_5),"средних квадратичных отклонения")
```
