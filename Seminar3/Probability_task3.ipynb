```python
from math import factorial
import math
import numpy as np
```

```python
#Задача 1
#Даны значения зарплат из выборки выпускников: 
#100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. 
#Посчитать (желательно без использования статистических методов наподобие std, var, mean) 
#среднее арифметическое, среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.
```

```python
#Решение
payments = np.array([100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150])
#Среднее арифметическое:
sum = 0
for i in payments:
    sum = sum + i
average = sum/payments.shape
print("Среднее арифметическое = ",average)
#Проверка встроенной функцией
print("Проверка:",np.average(payments))

#Смещенная дисперсия
summa = 0
for i in payments:
    elem = (i - average)**2
    summa+=elem
variance = summa/payments.shape
print("Смещенная дисперсия:",variance)
#Проверка встроенной функцией
print("Проверка:",np.var(payments))

#Несмещенная дисперсия
summa_2 = 0
for i in payments:
    elem_2 = (i - average)**2
    summa_2+=elem_2
variance_2 = summa/(payments.size-1)
print("Несмещенная дисперсия:",variance_2)
#Проверка встроенной функцией
print("Проверка:",np.var(payments,ddof=1))

#Cреднее квадратичное отклонение (смещенное)
standard_deviation = np.sqrt(variance)
print("Смещенное стандартное отклонение:",standard_deviation)
#Проверка встроенной функцией
print("Проверка:",np.std(payments))

#Cреднее квадратичное отклонение (несмещенное)
standard_deviation_2 = np.sqrt(variance_2)
print("Несмещенное стандартное отклонение:",standard_deviation_2)
#Проверка встроенной функцией
print("Проверка:",np.std(payments, ddof = 1))
```

```python
#Задача 2
#В первом ящике находится 8 мячей, из которых 5 - белые. Во втором ящике - 12 мячей, из которых 5 белых. 
#Из первого ящика вытаскивают случайным образом два мяча, из второго - 4. Какова вероятность того, что 3 мяча белые?
```

```python
#Решение
def combinations (a,b):
    amount_combinations = factorial(a)/(factorial(a-b)*factorial(b))
    return int(amount_combinations)

#Решение задачи состоит из трех возможных комбинаций:
# A1: из первого ящика достали оба белых, из второго 1 белый
# А2: из первого ящика достали один белый, из второго 2 белых
# А3: из первого ящика не достали ни одного белого, из второго 3 белых
# Общее число исходов рассчитывается одинаковым образом для всех трех ситуаций:
outcomes_first_box = combinations(8,2)
print("Число исходов для первого ящика:",outcomes_first_box)
outcomes_second_box = combinations(12,4)
print("Число исходов для второго ящика:",outcomes_second_box)

#Расчитаем вероятность для комбинации А1
probability_A1 = (combinations(5,2)/outcomes_first_box)*((combinations(5,1)*combinations(7,3))/outcomes_second_box)
print("Вероятность комбинации А1:", round((probability_A1),4))

#Расчитаем вероятность для комбинации А2
probability_A2 = ((combinations(3,1)*combinations(5,1))/outcomes_first_box)*((combinations(5,2)*combinations(7,2))/outcomes_second_box)
print("Вероятность комбинации А2:", round((probability_A2),4))

#Расчитаем вероятность для комбинации А3
probability_A3 = ((combinations(5,0)*combinations(3,2))/outcomes_first_box)*((combinations(5,3)*combinations(7,1))/outcomes_second_box)
print("Вероятность комбинации А3:", round((probability_A3),4))

#Вероятность наступления заданного события получается сложением вероятностей событий A1, A2, A3
probability = probability_A1 +probability_A2 +probability_A3
print("Вероятность, что 3 мяча белые:", round((probability),4))
print("Вероятность в процентах:", round((probability),4)*100,"%")
```

```python
#Задача 3
#На соревновании по биатлону один из трех спортсменов стреляет и попадает в мишень. 
#Вероятность попадания для первого спортсмена равна 0.9, 
#для второго — 0.8, 
#для третьего — 0.6. 
#Найти вероятность того, что выстрел произведен: 
#    a). первым спортсменом 
#    б). вторым спортсменом 
#    в). третьим спортсменом.
```

```python
#Решение
#Событие А - цель поражена
#Событие B1 - цель поражена первым спортсменом
#Событие B2 - цель поражена вторым спортсменом
#Событие B3 - цель поражена третьим спортсменом
#B1,B2,B3 = полная группа событий, P(B1) = P(B2) = P(B3)= 1/3
#Для решения задачи используется формула Байеса
#P(B|A) = (P(A|B)*P(B))/P(A)
#Рассчитываем вероятность:
#Вероятность того, что выстрел произведен первым спортсменом
probability_B1_A = (1/3*0.9)/((1/3*0.9)+(1/3*0.8)+(1/3*0.6))
print("Вероятность того, что выстрел произведен первым спортсменом",round((probability_B1_A),4))
print("Вероятность в процентах:", round((probability_B1_A),4)*100,"%")

#Вероятность того, что выстрел произведен вторым спортсменом
probability_B2_A = (1/3*0.8)/((1/3*0.9)+(1/3*0.8)+(1/3*0.6))
print("Вероятность того, что выстрел произведен вторым спортсменом",round((probability_B2_A),4))
print("Вероятность в процентах:", round((probability_B2_A),4)*100,"%")

#Вероятность того, что выстрел произведен третьим спортсменом
probability_B3_A = (1/3*0.6)/((1/3*0.9)+(1/3*0.8)+(1/3*0.6))
print("Вероятность того, что выстрел произведен третьим спортсменом",round((probability_B3_A),4))
print("Вероятность в процентах:", round((probability_B3_A),4)*100,"%")
```

```python
#Задача 4
#В университет на факультеты A и B поступило равное количество студентов, 
#а на факультет C студентов поступило столько же, сколько на A и B вместе. 
#Вероятность того, что студент факультета A сдаст первую сессию, равна 0.8. 
#Для студента факультета B эта вероятность равна 0.7, 
#а для студента факультета C - 0.9. Студент сдал первую сессию.
#Какова вероятность, что он учится: 
#    a). на факультете A 
#    б). на факультете  B
#    в). на факультете C?
```

```python
#Решение
#Событие "X" - студент сдал сессию
# События "Студент с факультета A или B или C" -образуют полную группу событий
#Вероятности P(A) = P(B) = 0.25, P(C) = 0.5 (на факультете "C" в два раза больше студентов)
#Для решения используется формула Байеса

#Вероятность, что студент с факультета A:
faculty_A_probability = (0.25*0.8)/((0.25*0.8)+(0.25*0.7)+(0.5*0.9))
print("Вероятность того, что студент с факультета A",round((faculty_A_probability),4))
print("Вероятность в процентах:", round((faculty_A_probability),4)*100,"%")

#Вероятность, что студент с факультета B:
faculty_B_probability = (0.25*0.7)/((0.25*0.8)+(0.25*0.7)+(0.5*0.9))
print("Вероятность того, что студент с факультета B",round((faculty_B_probability),4))
print("Вероятность в процентах:", round((faculty_B_probability),4)*100,"%")

#Вероятность, что студент с факультета A:
faculty_C_probability = (0.5*0.9)/((0.25*0.8)+(0.25*0.7)+(0.5*0.9))
print("Вероятность того, что студент с факультета C",round((faculty_C_probability),4))
print("Вероятность в процентах:", round((faculty_C_probability),4)*100,"%")
```

```python
#Задача 5
#Устройство состоит из трех деталей.
#Для первой детали вероятность выйти из строя в первый месяц равна 0.1, 
#для второй - 0.2,
#для третьей - 0.25.
#Какова вероятность того, что в первый месяц выйдут из строя:
#    а). все детали 
#    б). только две детали 
#    в). хотя бы одна деталь 
#    г). от одной до двух деталей?
```

```python
#Решение
#Вероятности выхода из строя деталей 1, 2 и 3
p_first_detail = 0.1
p_second_detail = 0.2
p_third_detail = 0.25
#Вероятности противоположных событий (деталь не выходит из строя):
q_first_detail = 1 - p_first_detail
q_second_detail = 1 - p_second_detail
q_third_detail = 1 - p_third_detail
```

```python
#Решение варианта "a"
#Вероятность выхода из строя всех трех деталей рассчитывается через произведение вероятностей
#(совместные события, наступают одновременно) 
probability_a = p_first_detail*p_second_detail*p_third_detail
print("Вероятность выхода из строя всех трех деталей",round((probability_a),4))
print("Вероятность в процентах:",round((probability_a),4)*100,"%")
```

```python
#Решение варианта "б"
#Для решения задачи выделим следующие комбинации:
# 1 - выходит из строя деталь 1 и 2, 3 работает
# 2 - выходит из строя 1 и 3, 2 работает
# 3 - выходит из строя 2 и з, 1 работает
#Вероятность комбинации №1
comb_1 = p_first_detail*p_second_detail*q_third_detail
#Вероятность комбинации №2
comb_2 = p_first_detail*q_second_detail*p_third_detail
#Вероятность комбинации №3
comb_3 = q_first_detail*p_second_detail*p_third_detail
#Расчет вероятности выхода из строя только 2 деталей:
probability_b = comb_1 + comb_2 + comb_3
print("Вероятность выхода из строя только двух деталей",round((probability_b),4))
print("Вероятность в процентах:",round((probability_b),4)*100,"%")
```

```python
#Решение варианта "в"
#Решение будет исходить "от обратного события" - не выйдет из строя ни одна деталь
inverse_probability = q_first_detail*q_second_detail*q_third_detail
probability_c = 1 - inverse_probability
print("Вероятность выхода из строя хотя бы одной детали",round((probability_c),4))
print("Вероятность в процентах:",round((probability_c),4)*100,"%")
```

```python
#Решение варианта "д"
#Для решения найдем комбинации выхода из строя одной детали:
# а) первая деталь вышла из строя , вторая и третья работают
# б) вторая деталь вышла из строя, первая и третья работает
# в) третья деталь вышла из строя, первая и вторая работает
comb_a = p_first_detail*q_second_detail*q_third_detail
comb_b = q_first_detail*p_second_detail*q_third_detail
comb_c = q_first_detail*q_second_detail*p_third_detail
probability_combinations = comb_a + comb_b + comb_c
print("Вероятность выхода из строя одной детали",round((probability_combinations),4))
print("Вероятность в процентах:",round((probability_combinations),4)*100,"%")
#Вероятность выхода из строя от одной до двух деталей будет рассчитана как сумма вероятностей 
#выхода из строя одной детали и выхода из строя двух деталей
probability_d = probability_combinations + probability_b 
print("Вероятность выхода из строя от одной до двух деталей",round((probability_d),4))
print("Вероятность в процентах:",round((probability_d),4)*100,"%")
```
