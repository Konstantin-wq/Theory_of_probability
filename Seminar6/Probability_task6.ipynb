```python
import numpy as np
from scipy import stats as st
```

```python
#Задача 1
#Известно, что генеральная совокупность распределена нормально
#со средним квадратическим отклонением, равным 16.
#Найти доверительный интервал для оценки математического ожидания a с надежностью 0.95,
#если выборочная средняя M = 80, а объем выборки n = 256.
```

```python
#Решение
n_1 = 256 # объем выборки
mu_1 = 80 # выборочное среднее
sigma_1 = 16 # среднее квадратичное отклонение
alpha_1 = 1.96 # Табличное значение Z критерия
#Расчитаем границы доверительного интервала
left_1 = mu_1 - alpha_1*(sigma_1/np.sqrt(n_1)) #Расчитаем левую границу
right_1 = mu_1 + alpha_1*(sigma_1/np.sqrt(n_1)) #Расчитаем правую границу
interval_1 = [left_1,right_1]
print("Доверительный интервал для оценки математического ожидания с надежностью 0,95:",interval_1)
```

```python
#Задача 2
#В результате 10 независимых измерений некоторой величины X, выполненных с одинаковой точностью,
#получены опытные данные:
#6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1
#Предполагая, что результаты измерений подчинены нормальному закону распределения вероятностей,
#оценить истинное значение величины X при помощи доверительного интервала, покрывающего это
#значение с доверительной вероятностью 0,95.
```

```python
#Решение
exp_data = np.array([6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1]) #Опытные данные

n_2 = len(exp_data) #Находим объем выборки
print("Объем выборки:",n_2)

mu_2 = np.mean(exp_data)#находим среднее по выборке
print("Среднее значение по выборке:",mu_2)

d_2 = np.var(exp_data,ddof=1) #находим несмещенную дисперсию
print("Несмещенная дисперсия:",d_2)

sigma_2 = np.sqrt(d_2) # среднее квадратичное отклонение
print("Cреднее квадратичное отклонение",sigma_2)

#В условии задачи не указано стандартное отклонение генеральной совокупности, поэтому расчитаем t-критерий
p_2 = 0.95 #доверительная вероятность
alpha_2 = round((1-p_2)/2,4) #уровень статистической значимости
t_crit_2 = st.t.ppf(1-alpha_2,len(exp_data)-1)
print("Значение t-критерия:",t_crit_2)

#Расчитаем границы доверительного интервала
left_2 = round(mu_2 - t_crit_2*(sigma_2/np.sqrt(n_2)),4) #Расчитаем левую границу
right_2 = round(mu_2 + t_crit_2*(sigma_2/np.sqrt(n_2)),4) #Расчитаем правую границу

interval_2 = [left_2,right_2]
print("Доверительный интервал:",interval_2)
```

```python
#Задача 3
#Рост дочерей 175, 167, 154, 174, 178, 148, 160, 167, 169, 170
#Рост матерей 178, 165, 165, 173, 168, 155, 160, 164, 178, 175

#Используя эти данные построить 95% доверительный интервал для разности среднего роста родителей и детей.
```

```python
#Решение
daughters_height = np.array([175, 167, 154, 174, 178, 148, 160, 167, 169, 170])
mothers_height = np.array([178, 165, 165, 173, 168, 155, 160, 164, 178, 175])

#Находим объемы выборок
n_daughters = len(daughters_height)
n_mothers = len(mothers_height)
print(f'Объем выборки по росту дочерей равен {n_daughters}, объем выборки по росту матерей равен {n_mothers}.')

#Находим средние значения по выборкам
mu_daughters = np.mean(daughters_height)
mu_mothers = np.mean(mothers_height)
print(f'Среднее значение по росту дочерей равно {mu_daughters}, среднее значение по росту матерей равно {mu_mothers}.')

#Находим несмещенные дисперсии для каждой из выборок
d_daughters = np.var(daughters_height, ddof =1) 
d_mothers =  np.var(mothers_height, ddof =1)
print(f'Значение дисперсии по росту дочерей равно {d_daughters},значение дисперсии по росту матерей равно {d_mothers}.')

#Находим значение t-критерия выборок
p_3 = 0.95
alpha_3 = round((1-p_3)/2,4) #уровень статистической значимости
t_crit_3 = st.t.ppf((1-alpha_3),2*(len(daughters_height)-1))
print("Значение t-критерия:",t_crit_3)

#Находим параметры необходимые для разности выборок
delta_D = (d_mothers + d_daughters)/2 # объединенная оценка дисперсий
print("Объединенная оценка дисперсий:",delta_D) 

delta_mu = mu_mothers - mu_daughters #разность средних 
print("Разность средних:",delta_mu)

se_3 = np.sqrt((delta_D/10 + delta_D/10)) #стандартная ошибка разности средних
print("Стандартная ошибка разности средних:",se_3)

#Вычисляем доверительный интервал
left_3 = round((delta_mu - t_crit_3 *se_3),4)
right_3 = round((delta_mu + t_crit_3 *se_3),4)
interval_3 = [left_3,right_3]
print("Доверительный интервал:",interval_3)
```
