```python
#Задача 2
#Проведите тест гипотезы. 
#Утверждается, что шарики для подшипников, изготовленные автоматическим станком, имеют средний диаметр 17 мм. 
#Используя односторонний критерий с α =0,05, проверить эту гипотезу, 
#если в выборке из n=100 шариков средний диаметр оказался равным 17.5 мм, 
#а дисперсия известна и равна 4 кв. мм.
```

```python
#Решение
#Для решения задачи выбираем z-критерий 
#Формулируем нулевую и альтернативную гипотезы:
#Нулевая гипотеза Н_0 = все подшипники имеют диаметр 17 мм (mu = mu_0)
#Альтернативная гипотеза H_1 = средний размер подшипника >17мм (mu > mu_0)
#Формула для расчета z-критерия: z = (X_avg - mu)/(sigma/sqrt(n))
n = 100 #Размер выборки
mu = 17 #Среднее по генеральной совокупности
x_avg = 17.5 #среднее по выборке
d_x = 4 #Дисперсия
alpha = 0.05 # Уровень статистической значимости
sigma_2 = np.sqrt(d_x) #Находим стандартное отклонение, как квадратный корень из дисперсии
z_value_2  = (x_avg - mu)/(sigma_2/np.sqrt(n))
print("Наблюдаемое значение Z - критерия:",z_value_2)
#По табличным значениям Z критерия находим приблизительное:
z_table = 1.645
print("Табличное значение",z_table)
#Функция сравнения значения наблюдаемого и табличного значения
def compare(calc_value,table_value):
    if calc_value>table_value:
        print("Отвергаем нулевую гипотезу")
    else:
        print("Принимаем нулевую гипотезу")
#Ответ к задаче
compare(z_value_2,z_table)

#Проверочный расчет встроенными функциями
z_2=st.norm.ppf(0.95)
print("Расчет z-критерия при помощи функции ",z_2)
compare(z_value_2,z_2)
```

```python
#Задача 3
#Проведите тест гипотезы. Продавец утверждает, что средний вес пачки печенья составляет 200 г.
#Из партии извлечена выборка из 10 пачек. Вес каждой пачки составляет:
#202, 203, 199, 197, 195, 201, 200, 204, 194, 190.
#Известно, что их веса распределены нормально.
#Верно ли утверждение продавца, если учитывать, что доверительная вероятность равна 99%? (Провести двусторонний тест.)
```

```python
pack_weight = np.array([202, 203, 199, 197, 195, 201, 200, 204, 194, 190])

p = 0.99 #Доверительная вероятность
mu = 200# Среднее по ГС

x_avg = np.mean(pack_weight) #Находим среднее по выборке
print("Среднее значение по выборке:",x_avg)

n = len(pack_weight) #Находим размер выборки
print("Размер выборки:",n)

d_x = np.var(pack_weight,ddof=1) #Расчет несмещенной дисперсии
print("Несмещенная дисперсия:",d_x)    

#В условии задачи не указано стандартное отклонение генеральной совокупности, поэтому применим расчет t-критерия
t_crit = (x_avg - mu)/(np.sqrt(d_x)/np.sqrt(n))
print("Значение t-критерия:",t_crit)

#Для двустороннего теста вычислим значение статистической значимости
alpha = round((1-p)/2,4)
print("Значение статистической значимости:",round((1-p)/2,4))

#Находим табличное соотвтетствующее табличное значение 
t_table = 3.250
print("Табличное значение:",t_table)
#Проверяем значение встроенной функцией
t_table_func = st.t.ppf(1-alpha,len(pack_weight)-1)
print("Значение полученное при помощи встроенной функции:",t_table_func)

compare(abs(t_crit),t_table)
```
